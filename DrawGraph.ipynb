```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
# 创建一个图形和一个轴
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r-', animated=True)

# 设置图形的界限
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1, 1)

# 初始化函数，用于设置图形的初始状态
def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

# 更新函数，用于更新图形
def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

# 创建动画
ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)

plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

# 创建实时绘制横纵轴变量
x = []
y = []
z = []

# 创建绘制实时损失的动态窗口
plt.ion()

# 创建循环
for i in range(5):
    x.append(i)	# 添加i到x轴的数据中
    y.append(i**2)	# 添加i的平方到y轴的数据中
    z.append(i**3)
    plt.clf()  # 清除之前画的图
    plt.plot(x, y,x,z)  # 画出当前x列表和y列表中的值的图形
    plt.pause(0.001)  # 暂停一段时间，不然画的太快会卡住显示不出来
    plt.ioff()  # 关闭画图窗口
```
